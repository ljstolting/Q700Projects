```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as spsg
import statistics as stat

##create bursting 3-neuron CTRNN

def sigmoid(x):
    return 1/(1+np.exp(-x))

class CTRNN():

    def __init__(self,duration,dt):
        size = 2
        self.Size = size                        # number of neurons in the circuit
        self.States = np.ones(size)            # state of the neurons
        self.TimeConstants = np.zeros(size)      # time-constant for each neuron
        self.invTimeConstants = 1.0/self.TimeConstants
        self.Biases = np.zeros(size)            # bias for each neuron
        self.Weights = np.zeros((size,size))    # connection weight for each pair of neurons
        self.Outputs = np.zeros(size)           # neuron outputs
        self.Inputs = np.zeros(size)            # external input to each neuron
        self.time = np.arange(0.0,duration,dt)
        self.TimeSeries = np.zeros((len(self.time),3))
        self.dt = dt
        

    def setInputs(self,ext_inputs):
        self.Inputs = ext_inputs
    
    def setWeights(self, weights):
        self.Weights =  weights

    def setBiases(self, biases):
        self.Biases =  biases

    def setTimeConstants(self, timeconstants):
        self.TimeConstants =  timeconstants
        self.invTimeConstants = 1.0/self.TimeConstants

    def randomizeParameters(self):
        self.Weights = np.random.uniform(-10,10,size=(self.Size,self.Size))
        self.Biases = np.random.uniform(-10,10,size=(self.Size))
        self.TimeConstants = np.random.uniform(0.1,5.0,size=(self.Size))
        self.invTimeConstants = 1.0/self.TimeConstants

    def initializeState(self, s):
        self.States = s
        self.Outputs = sigmoid(self.States+self.Biases)

    def step(self):
        netinput = self.Inputs + np.dot(self.Weights.T, self.Outputs)
        self.States += self.dt * (self.invTimeConstants*(-self.States+netinput))
        self.Outputs = sigmoid(self.States+self.Biases)
        
        
    def run(self):
        for stepnum in range(len(self.time)):
            self.step()
            self.TimeSeries[stepnum] = self.Outputs
        groupedoutputs = self.TimeSeries.T
        self.timederivs = np.zeros((3,len(self.time)))
        for i in range(3):
            self.timederivs[i] = np.gradient(groupedoutputs[i])
            
    def plot(self):
        plt.rcParams["figure.figsize"] = (10,10)
        ax1 = plt.subplot(211)
        ax2 = plt.subplot(212)
        lab = ["BS","FS","FT"]
        for i in range(3):
            ax1.plot(self.time,self.TimeSeries[:,i],label=lab[i])
        for i in range(3):
            ax2.plot(self.time,self.timederivs[i],label=lab[i])
        ax1.set_xlabel("Time")
        ax1.set_ylabel("Spike Frequency")
        ax1.set_title("Neural activity")
        ax2.set_xlabel("Time")
        ax2.set_ylabel("Time Derivative")
        ax2.set_title("Rate of Change")
        ax1.legend()
        ax2.legend()
        plt.show()
        
    def calc_phase(self): #Calculate beginning of each burst for each neuron by finding peaks of time derivative 
        num_cycles = len(spsg.find_peaks(self.timederivs[0],prominence = .0005)[0])-1   #minus one b/c disregarding last cycle
        #print(num_cycles)
        onset_times = np.zeros((3,num_cycles))

        for i in range(3):
            onset_times[i]= spsg.find_peaks(self.timederivs[i],prominence = .0005)[0][:num_cycles]
            onset_times[i] = onset_times[i]*self.dt
            
        #self.onset_times = onset_times
        self.period = stat.mean([onset_times[0,1]-onset_times[0,0],onset_times[1,1]-onset_times[1,0],onset_times[2,1]-onset_times[2,0]])
        shifted_onset_times = np.zeros((3,num_cycles))
        
        for i in range(1,3):
            if onset_times[i,0]<onset_times[0,0]:    #if the first burst of a subsequent neuron type occurs before the first burst of neuron BS
                onset_times[i] = np.concatenate((np.array([0]),onset_times[i,1:]))   #replace the first element with a 0 to shift its phase
                
        onset_times = np.delete(onset_times,0,1)     #and then delete the first column
        #print('hi')        

        for j in range(len(onset_times[0])):
            for i in range(3):
                shifted_onset_times[i,j] = onset_times[i,j] - onset_times[0,j]
        #print('hi')

        phases = shifted_onset_times/self.period

        self.avg_phases = np.average(phases,axis=1)
        #print(self.avg_phases)

        
```
