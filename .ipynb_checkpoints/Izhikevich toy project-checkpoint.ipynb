```python
import matplotlib.pyplot as plt
import numpy as np

# Global Parameters
duration = 1000
maxinput = 100
stepsize = 0.1
cutofftime = 50

v0=-65
detailplotinputs = [5,10,20,40,60,80]
#detailplotinputs = list(np.arange(0.0,maxinput+1,(maxinput)/5))  #choose 6 input values to plot the detailed time series for
detailplot = [[] for x in range(6)]
detailISI = [[] for x in range(6)]
class IzhikevichNeuron:

    def __init__(self, v, a, b, c, d):
        self.Voltage = v
        self.u = b*v
        self.a = a
        self.b = b
        self.c = c
        self.d = d
        self.spikecount = 0
        self.intervaldict = {}
        self.intervaltimecounter = 0.0
        
    def step(self,dt,Input,time,detailplotis):
        dVdt = (0.04 * (self.Voltage**2)) + (5 * self.Voltage) + 140.0 - self.u + Input
        dudt = self.a * ((self.b * self.Voltage) - self.u)
        self.Voltage += dt * dVdt
        self.u += dt * dudt
        if self.Voltage >= 30.0:
            self.Voltage = self.c
            self.u = self.u + self.d
            self.spikecount += 1
            if time >= cutofftime+self.intervaltimecounter: #if the last reference spike occurred after the start of counting
                if self.intervaltimecounter in self.intervaldict:
                    self.intervaldict[self.intervaltimecounter] += 1
                else:
                    self.intervaldict[self.intervaltimecounter] = 1
                #print(self.intervaldict)

            self.intervaltimecounter = 0.0
        else:
            self.intervaltimecounter += dt
            self.intervaltimecounter = round(self.intervaltimecounter, 4)
        if Input in detailplotis:
            detailplot[detailplotinputs.index(Input)].append(self.Voltage)
            #detailISI[detailplotinputs.index(Input)].append(self.intervaltimecounter)
            
a=0.02
b=0.2
c=-50
d=6.55


timeseries = np.arange(0.0,duration+stepsize,stepsize)
inputs = np.arange(0.0,maxinput+1,1.0)
count = np.zeros(len(inputs))
k = 0

fig1, f1_axs = plt.subplots(ncols=3, nrows=3)
gs = f1_axs[1, 2].get_gridspec()
# remove the underlying axes
for ax in f1_axs[2,:]:
    ax.remove()
ISIs = fig1.add_subplot(gs[2,:])

i=0
for ax in f1_axs.flat[0:6]:
    ax.set_xlabel('time(s)',fontsize=20)
    ax.set_ylabel('Voltage(mV)',fontsize=20)
    ax.set_title('Input=' + str(detailplotinputs[i]),fontsize=20)
    i+=1
ISIs.set_xlabel("Inputs(mV)",fontsize=20)
ISIs.set_ylabel("ISI(s)",fontsize=20)
ISIs.set_title('ISIs across Inputs',fontsize=20)


fig1.set(figheight=20,figwidth=20)

for inp in inputs:
    # Create neuron
    n = IzhikevichNeuron(v0,a,b,c,d)
    # Run simulation
    for t in timeseries:
        n.step(stepsize,inp,t,detailplotinputs)
    count[k] = n.spikecount
    k += 1
    for key in n.intervaldict:
        if inp in detailplotinputs:
            ISIs.plot(inp,key,'ro')
        else:
            ISIs.plot(inp,key,'ko')


            
for i in range(6):
    f1_axs.flat[i].plot(timeseries,detailplot[i])
```

```python
import statistics as stat
inputs = np.arange(15.0,maxinput+1,1.0)     #must start after I=15 for this data set (regular spiking at low input voltages means no variance)
def switchcalc(d,inputs):
    variances = []
    for inp in inputs:
        #print(inp)
        ISIlist = []
    # Create neuron
        neu = IzhikevichNeuron(v0,a,b,c,d)
    # Run simulation
        for t in timeseries:
            neu.step(stepsize,inp,t,detailplotinputs)
        #print(neu.intervaldict)
        for key in neu.intervaldict:
            i=neu.intervaldict[key]
            for j in range(i):
                ISIlist.append(key)
                j += 1
       # print(ISIlist)
        if len(ISIlist) > 0:
            vari = stat.variance(ISIlist)
            #print(vari)
            variances.append((inp,vari))
            #print(variances)
            if vari <= 4:
                return inp
        else:
            pass
```

```python
switchmarks = []
for d in np.arange(.05,8.06,.5):
    switchmarks.append(switchcalc(d,inputs))
```

```python
switchcalc(6.5,inputs)  #in this case, 6.5 is the value of d we investigate
variances2 = zip(*variances)
plt.plot(*variances2)
plt.title('Variance of ISIs vs. Input')
plt.xlabel('Input (mV)')
plt.ylabel('Variance (s^2)')
```

```python
dseries = np.arange(.05,8.06,.5)
plt.plot(dseries,switchmarks)
plt.title('Switching Input vs. d value')
plt.xlabel('d')
plt.ylabel('Threshold Input (mV)')
```
